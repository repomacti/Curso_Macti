**Teorema Fundamental del Cálculo** (primera parte)

Sea $f$ una función continua en el intervalo $[a,b]$, entonces la función $F$ definida como

$$F(x)=\int_{a}^{x}f(t)dt \;\;\; a \le x \le b$$

es continua en el intervalo $[a,b]$ y diferenciable en $(a,b)$ y $F^\prime(x)=f(x)$.
    
</div>

Básicamente este teorema dice que las operaciones de derivación y de integración son operaciones inversas. Entonces si calculamos la integral de $f(x)$ deberemos obtener $F(x)$.

**Cálculo del gradiente**

Definición del [Gradiente](https://es.wikipedia.org/wiki/Gradiente):
<blockquote>
En análisis matemático, particularmente en cálculo vectorial, el gradiente o vector gradiente de un campo escalar ${\displaystyle f:\mathbb {R} \longrightarrow \mathbb {R} ^{n}}$ es un campo vectorial, denotado $\nabla f$. El vector gradiente de $f$ evaluado en un punto genérico $x$ del dominio de $f$ indica la dirección en la cual el campo $f$ varía más rápidamente y su módulo representa el ritmo de variación de $f$ en la dirección de dicho vector gradiente.

La generalización del concepto de gradiente para funciones vectoriales de varias variables es el concepto de matriz jacobiana.
</blockquote>

Si $f$ es una función de varias variables $x, y, z, \dots$, entonces su gradiente se denota por:

$$
\nabla f = \left[
\begin{array}{c}
\frac{\partial f}{\partial x} \\
\frac{\partial f}{\partial y} \\
\frac{\partial f}{\partial z} \\
\vdots
\end{array}
\right]
$$
    
</div>

**Ley de decaimiento radioactivo**. 
La masa de una substancia radioactiva decae a una razón que es proporcional a la cantidad de masa que está presente. Si $y(t)$ expresa la cantidad de substancia  en el tiempo $t$, entonces esta ley se expresa como:

$$
\begin{eqnarray*}
\frac{d y(t)}{d t} & = & - \lambda y(t), \qquad \text{ para }  a < t  < b \\
y(a) & = & \alpha \qquad \text{(condición inicial)}
\end{eqnarray*}
$$

donde $\alpha$ representa la cantidad  de susbtancia inicial y $\lambda > 0$.

1. Resolver la ecuación de manera general.
2. Resolver la ecuación con la condición inicial.
2. Resolver para los siguientes valores: $(a,b) = (0,20)$, $\lambda = 1.5$, $\alpha = 20$.
3. Graficar el resultado de 2. con `sympy.plot()`.
4. Graficar el resultado de 2. con `matplotlib`.
</font>
    
</div>

### 1. Solución general.

```python
# 1.

# Definimos los símbolos necesarios t, y(t), λ y α
t = sympy.Symbol('t')
y = sympy.Function('y')
λ = sympy.symbols('λ', real=True)
α = sympy.symbols('α', real=True)
```

```python
# Definimos la función f(t, y)
f = -λ * y(t)
f
```

```python
# Definimos la ecuación diferencial
edo = sympy.Eq(y(t).diff(t), f)
edo
```

```python
# Obtenemos la solución de manera general.
sympy.dsolve(edo, y(t))
```

### 2. Solución con condición inicial.

```python
# Definimos la condición inicial
ics = {y(0): α}
```

```python
# Resolvemos con la condición inicial
sol = sympy.dsolve(edo, y(t), ics=ics)
sol
```

### 3. Solución particular.

```python
# Resolvemos con la condición inicial particular
sol = sympy.dsolve(edo, y(t), ics = {y(0): 20})
sol
```

```python
# Obtenemos el lado izquierdo de la ecuación
sol.lhs
```

```python
# Obtenemos el lado derecho de la ecuación
sol.rhs
```

```python
# Substituimos el valor de λ = 1.5
sol_e = sol.rhs.subs(λ, 1.5)
sol_e
```

### 4. Gráfica con sympy.

```python
ph = sympy.plot(sol_e, (t, 0, 20), show = False)
ph.size = (4,2)
ph.show()
```

### 5. Gráfica con matplotlib.

```python
# Función para generar datos
fg = sympy.lambdify(t, sol.rhs.subs(λ, 1.5))

# Dominio de la solución
tg = np.linspace(0,10,100)

# Evaluación
fg(tc)
```

```python
# Gráfica con matplotlib
plt.plot(tc, fg(tc))
```
